# Train Hardcoded & Baseline Models

tip - always create  a simple hardcoded odel to establish the mininmum scre that any proper ML model should beat

* Hardcoded Model : always predict ave fare
* Baseline model : Linear Regression

for evaluation use RMSE

```python
import numpy as np
```

```python

class MeanRegressor :
  def fit(self, input, targets):
    self.mean = targets.mean()

  def predict(self, inputs):
    return np.full(inputs.shape[0], self.mean)
```

```python
mean_model = MeanRegressor()
mean_model.fit(train_inputs, train_target)
mean_model.mean
```

```python
train_pred = mean_model.predict(train_inputs)
```

```python
train_pred
```

```python
val_pred = mean_model.predict(val_inputs)
```

```python
from sklearn.metrics import mean_squared_error
```

```python

def rmse(targets,preds):
  return np.sqrt(mean_squared_error(targets,preds))
```

```python
train_rmse = rmse(train_target, train_pred)
train_rmse
```

```python
val_rmse = rmse(val_target, val_pred)
val_rmse
```

```python
def add_dateparts(df,col):
  df[col + 'year'] = df[col].dt.year
  df[col + 'month'] = df[col].dt.month
  df[col + 'day'] = df[col].dt.day
  df[col + 'weekday'] = df[col].dt.weekday
  df[col + 'hour'] = df[col].dt.hour
```

#Distance between two coordinates

```python
import math
```

```python
def haversine_np(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)

    All args must be of equal length.
    """
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat / 2.0) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0) ** 2
    c = 2 * np.arcsin(np.sqrt(a))

    km = 6367 * c
    return km
```

```python
def add_trip_distance(df):
  df['trip_distance'] = haversine_np(df['pickup_latitude'],
                                  df['pickup_longitude'],
                                  df['dropoff_latitude'],
                                  df['dropoff_longitude'])
```

```python
add_trip_distance(train_df)
add_trip_distance(val_df)
add_trip_distance(test_df)
```
