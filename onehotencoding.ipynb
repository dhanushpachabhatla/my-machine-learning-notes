```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.sample(4)
```

#One hot encoding using Pandas

```python
dummies = pd.get_dummies(df,columns = ['fuel', 'owner'])
dummies
```

#K-1 OneHotEncoding

```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
```

#ONE with Top Categories (most repeating ones and others )

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon')).sample(5)
```
