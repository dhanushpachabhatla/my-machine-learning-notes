#Sparsed Tensors

A sparse tensor is just a way of saving memory when most values are zero. Instead of storing all the zeroes, TensorFlow just stores:

* where the non-zero values are (their indices)

* what the non-zero values are

* the full shape of the tensor

```python
tensor_sparsed = tf.sparse.SparseTensor(indices=[[0,0],[1,2]],values=[1,2],dense_shape=[3,4])
print(tensor_sparsed)
```

```python
print(tf.sparse.to_dense(tensor_sparsed))
```
